```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
```

 Paul Erdos coauthors graph
Paul Erdős was a Hungarian mathematician who wrote many papers in diverse fields of mathematics. Coauthors of Erdős are said to have an "Erdos number" 1; those who collaborated with them hold an Erdős number 2, and so on. Erdős himself has an Erdős number 0.

```python
# load Erdos adjacency matrix and Erdos'coauthors
A = np.load('Erdos_adjacency.npy')
coauthors = np.load('Erdos_coauthors.npy')
```

The matrix 
A
 is the adjacency matrix of the graph of collaborations between authors with Erdős number 1. The graph have Erdos's coauthors as its vertices

```python
# visualize the adjacency matrix A
plt.figure(figsize=(8,8))
plt.spy(A,marker='o',markersize=7)
plt.axis(False)
```

```python
# laplacian L=D-A
L= np.diag( np.sum(A, axis=0))-A
#Eigen values,vectors
D,V = eigh(L)
c1=np.where(V[:,1]>=0)[0]
c2=np.where(V[:,1]<=0)[0]
n1=len(c1)
n2=len(c2)
#permutation
p=np.hstack((c1,c2)).flatten()
A_per=A[p][:,p]
```

```python
# visualize the adjacency matrix A
plt.figure(figsize=(8,8))
plt.spy(A_per,marker='o',markersize=7)
plt.axis(False)
```

```python
n=A.shape[0]
A1=A_per[0:n1,0:n1]
A2=A_per[n1:n,n1:n]
```

```python
# laplacian L=D-A
L1= np.diag( np.sum(A1, axis=0))-A1
#Eigen values,vectors
D1,V1 = eigh(L1)
c11=np.where(V1[:,1]>=0)[0]
c12=np.where(V1[:,1]<=0)[0]
n11=len(c11)
n12=len(c12)
#permutation
p1=np.hstack((c11,c12)).flatten()
```

```python
# laplacian L=D-A
L2= np.diag( np.sum(A2, axis=0))-A2
#Eigen values,vectors
D2,V2 = eigh(L2)
c21=np.where(V2[:,1]>=0)[0]
c22=np.where(V2[:,1]<=0)[0]
n21=len(c21)
n22=len(c22)
#permutation
p2=np.hstack((c21,c22)).flatten()
```

```python
# combine permutations
pp = p.copy()
pp[0:n1] = pp[0:n1][p1]
pp[n1:n] = pp[n1:n][p2]
```

```python
A_per = A[p][:,p]
A_per_per = A[pp][:,pp]

plt.figure(figsize=(8,8))
plt.spy(A_per_per, marker='o', markersize=7)
plt.axis(False)
```
