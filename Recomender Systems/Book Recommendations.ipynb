# Book Recommendations

The [Book-Crossings dataset](http://www2.informatik.uni-freiburg.de/~cziegler/BX/) is a book ratings dataset compiled by Cai-Nicolas Ziegler from the [Book-Crossing community](https://www.bookcrossing.com). 
It contains 1.1 million ratings of 270,000 books by 90,000 users. 
The ratings are on a scale from 1 to 10.

Click [here](http://www2.informatik.uni-freiburg.de/~cziegler/BX/BX-CSV-Dump.zip) to download the dataset.
The data consists of three files: ratings (BX-Book-Ratings.csv), books info (BX-Books.csv), and users info (BX-Users.csv)

## Part 4

Use

- (a) Jaccard similarity $$ J(A,B) = \frac{|A\cap B|}{|A\cup B|} $$ 
- (b) serendipity similarity $$S(A,B) = \frac{|A\cap B|}{|B|}$$
- (c) overlap similarity $$O(A,B) = \frac{|A\cap B|}{\min\{|A|,|B|\}}$$
- (d) correlation similarity, and
- (e) cosine similarity

to find the top-10 book recomendations

```python
#Jaccard similarity
def Jaccard_sim(set_B):
    set_A = book_sets[book]
    return len(set_A.intersection(set_B))/len(set_A.union(set_B))
```

```python
# top 20 recommendations
book_sets.apply(Jaccard_sim).sort_values(ascending=False).head(20)
```

```python
def Serendipity_sim(set_B):
    set_A = book_sets[book]
    return len(set_A.intersection(set_B))/len(set_B)
```

```python
book_sets.apply(Serendipity_sim).sort_values(ascending=False).head(20)
```

```python
def Overlap_sim(set_B):
    set_A = book_sets[book]
    return len(set_A.intersection(set_B))/min(len(set_A),len(set_B))
```

```python
book_sets.apply(Overlap_sim).sort_values(ascending=False).head(20)
```

```python
# the ratings matrix
ratings_matrix = ratings.pivot_table(index='User-ID',columns='Book-Title',values='Book-Rating')
ratings_matrix
```

```python
def correlation_sim(col_B):
    col_A = ratings_matrix.loc[:,book]
    
    # means
    mean_A = np.nansum(col_A)/np.sum(~np.isnan(col_A))
    mean_B = np.nansum(col_B)/np.sum(~np.isnan(col_B))
    
    norm_A = np.sqrt(np.nansum((col_A-mean_A)**2))
    norm_B = np.sqrt(np.nansum((col_B-mean_B)**2))
    return np.nansum((col_A-mean_A)*(col_B-mean_B))/(norm_A*norm_B)
```

```python
ratings_matrix.apply(correlation_sim).sort_values(ascending=False).head(20)
```

```python
def cosine_sim(col_B):
    col_A = ratings_matrix.loc[:,book]
    norm_A = np.sqrt(np.nansum(col_A**2))
    norm_B = np.sqrt(np.nansum(col_B**2))
    return np.nansum(col_A*col_B)/(norm_A*norm_B)
```

```python
ratings_matrix.apply(cosine_sim).sort_values(ascending=False).head(20)
```
